```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transdata = pd.read_csv(r'C:\Users\jhoan\Downloads\transactions_data.csv')
transdata.head(10)
                        
```

```python
#Delete Unnecessary Column -  Unnamed: 0
del transdata['Unnamed: 0']
transdata
```

```python
#Check data info
transdata.info
```

```python
#Check duplicates
transdata.duplicated().sum()
```

```python
#Delete unnecessary column
del transdata['Transaction_ID']
transdata
```

```python
#Change the date format
transdata['Transaction_Date'] = pd.to_datetime(transdata['Transaction_Date'], format='mixed')
transdata
```

```python
#Change quantity, product and quantity null values to mode value
transdata['Company_ID'] = pd.to_numeric(transdata['Company_ID'], errors='coerce')
median_companyID = transdata['Company_ID'].median()
transdata['Company_ID'].fillna(median_companyID, inplace=True)
transdata['Product_ID'] = pd.to_numeric(transdata['Product_ID'], errors='coerce')
median_productID = transdata['Product_ID'].median()
transdata['Product_ID'].fillna(median_productID, inplace=True)
transdata['Quantity'] = pd.to_numeric(transdata['Quantity'], errors='coerce')
median_Quantity = transdata['Quantity'].median()
transdata['Quantity'].fillna(median_Quantity, inplace=True)
transdata
```

```python
#Fill NA values in Product Price with Total Cost
transdata = transdata.fillna(method='bfill',axis=1)
transdata
```

```python
#Fill NA values in Total Cost with Product Price
transdata = transdata.fillna(method='ffill',axis=1)
transdata
```
