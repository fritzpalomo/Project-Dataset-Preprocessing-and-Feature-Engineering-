```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transdata =pd.read_csv('cleaned_customer_transactions.csv')
transdata.head(10)
```

```python
#Adding date segmentation features
transdata['Transaction_Date'] = pd.to_datetime(transdata['Transaction_Date'], format='mixed')
transdata['year'] = transdata['Transaction_Date'].dt.year
transdata['month'] = transdata['Transaction_Date'].dt.month
transdata['day_of_week'] = transdata['Transaction_Date'].dt.weekday
transdata
```

```python
#Adding a column to indicate if the sale happened during weekend/weekday
transdata['is_weekend'] = transdata['day_of_week'].isin([5, 6]).astype(int)
transdata
```

```python
#Adding a column to identify quarter
transdata['quarter'] = transdata['Transaction_Date'].dt.quarter
transdata
```

```python
del transdata['product_popularity'], transdata['monthly_sales']
transdata
```

```python
transdata.to_csv('feature_engineering_customer_transactions.csv')
print('feature_engineering_customer_transactions.csv')
```
