```python
import pandas as pd

transdata = pd.read_csv(r'C:\Users\jhoan\Downloads\transactions_data.csv')
transdata.head(10)
```

```python
del transdata['Unnamed: 0'], transdata['Transaction_ID']
transdata
```

```python
transdata.describe()
```

```python
transdata.isnull().sum()
```

```python
transdata.duplicated().sum()
```

```python
transdata['Company_ID'] = transdata['Company_ID'].fillna('Unknown')
transdata['Product_ID'] = transdata['Product_ID'].fillna('Unknown')
transdata['Quantity'] = transdata['Quantity'].fillna(0.00)
transdata['Product_Price'] = transdata['Product_Price'].fillna(0.00)
transdata['Total_Cost'] = transdata['Total_Cost'].fillna(0.00)
transdata
```

```python
transdata.isnull().sum()
```

```python
transdata['Transaction_Date'] = pd.to_datetime(transdata['Transaction_Date'], format='mixed')
transdata
```

```python
transdata.to_csv('cleaned_transactions_data.csv', index=False)
print('cleaned_transactions_data.csv')
```
